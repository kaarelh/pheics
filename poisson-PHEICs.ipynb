```python
import math

current=2022.5
predictedyear=2050
diff=predictedyear+1-current #finds time (in years) from now until end of predictedyear
rate=3/13 #PHEICs will be modeled as a https://en.wikipedia.org/wiki/Poisson_point_process#Interpreted_as_a_counting_process with this lambda 

problist=[] #list to store probabilities of i PHEICs
for i in range(100):
    prob=(rate*diff)**i/math.factorial(i)*math.exp(-rate*diff)
    problist.append(prob)
print(problist)
psum=0
c5=0
c25=0
c50=0
c75=0
c95=0
for i in range(len(problist)):
    psum+=problist[i]
    if psum>0.05:
        if c5==0:
            print(i)
        c5=1
    if psum>0.25:
        if c25==0:
            print(i)
        c25=1
    if psum>0.5:
        if c50==0:
            print(i)
        c50=1
    if psum>0.75:
        if c75==0:
            print(i)
        c75=1
    if psum>0.95:
        if c95==0:
            print(i)
        c95=1
    
         
    
    
```
